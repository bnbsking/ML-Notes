### Read

```python
import cv2; 
import matplotlib.pyplot as plt
import PIL

img1 = cv2.imread("groundhog.jpeg")
print( type(img1), img1.shape )

img2 = plt.imread("groundhog.jpeg")
print( type(img2), img2.shape )

img3 = PIL.Image.open("groundhog.jpeg")
print( type(img3), img3.size )
```

### Show

```python
cv2.imshow("windowName",img1); cv2.waitKey(0); cv2.destroyAllWindows() # popup # press 0 to exit instead of clicking escape
```

```python
# recommended
plt.imshow(img1); plt.show()
plt.imshow(img1[:,:,::-1]); plt.show()
plt.imshow(img2); plt.show()
plt.imshow(img3); plt.show()
```

```python
img3.show() # popup
```

```python
import numpy as np
img3np = np.array(img3)
plt.imshow(img3np); plt.show()
```

### Save

```python
cv2.imwrite("test1.jpg", img1)
plt.imshow(img2); plt.savefig("test2.jpg")
img3.save("test3.jpg")
```
