### Self-writing linear regression practice

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([1,2,3])
y = np.array([1,2,3])

class Model:
    def __init__(self,a=0,b=0):
        self.a, self.b = a, b
    def predict(self,x):
        return self.a*x+self.b
    
def mse(y,yPred):
    return sum( (y-yPred)**2 ) / len(y)

def grad(model, X, y, h=1e-4):
    mse0  = mse(y,model.predict(X))
    
    model.a += h
    grad_a = (mse(y,model.predict(X)) - mse0)/h
    model.a -= h
    
    model.b += h
    grad_b = (mse(y,model.predict(X)) - mse0)/h
    model.b -= h
    
    return grad_a, grad_b
```

```python
model = Model()
lossL = []
lr = 1e-1

for i in range(350):
    loss = mse(y,model.predict(X))
    g = grad(model, X, y)
    model.a = model.a - lr*g[0]
    model.b = model.b - lr*g[1]
    lossL.append(loss)
    if i%10==0:
        print(i, round(model.a,3), round(model.b,3), round(loss,3))
        
print(model.predict(5))
```

### Parabola

```python
def parabola(X):
    return X[0]**2 + (X[1]-1)**2

def grad(X, h=1e-4):
    l0 = parabola(X)
    
    X[0] += h
    grad_a = ( parabola(X) - l0 )/h
    X[0] -= h
    
    X[1] += h
    grad_b = ( parabola(X) - l0 )/h
    X[1] -= h
    
    return grad_a, grad_b
```
