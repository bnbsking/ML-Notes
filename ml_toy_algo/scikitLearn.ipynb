```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html
from sklearn.linear_model import LinearRegression
X = np.array([[-1], [-2], [-3], [1], [2], [3]])
y = np.dot(X, np.array([2])) + 3 # y=2x+3
print(X.shape, y.shape)
reg = LinearRegression().fit(X, y)

print(reg.score(X,y)) # R^2 # 0~1
print(reg.predict(np.array([[2.5]])))
plt.scatter(X, y)
plt.scatter([2.5], reg.predict(np.array([[2.5]])))
plt.show()
```

```python
X = [-1,-2,-3,1,2,3]
X = np.array([ [x**2,x,1] for x in X ])
y = np.dot(X, np.array([1,0,1])).reshape(6) # y=x*2+1
print(X.shape, y.shape)
reg = LinearRegression().fit(X, y)

print(reg.score(X,y)) # R^2 # 0~1
print(reg.predict(np.array([[2.5**2,2.5,1]])))
plt.scatter(X[:,1], y)
plt.scatter([2.5], reg.predict(np.array([[2.5**2,2.5,1]])))
plt.show()
```

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html
from sklearn.svm import SVC
X = np.array([[-1,-1], [-2,-1], [-3,-2], [1,1], [2,1], [3,2], [-2,1], [-3,2], [-1,2]])
y = [0,0,0,1,1,1,2,2,2]
svc = SVC().fit(X, y) # kernel{‘linear’, ‘poly’, ‘rbf’, ‘sigmoid’, ‘precomputed’}, default=’rbf’

print(svc.predict([[2., 2.]]))
plt.scatter(X[:3,0], X[:3,1]) 
plt.scatter(X[3:6,0], X[3:6,1])
plt.scatter(X[6:,0], X[6:,1])
plt.scatter([2], [2], marker='x')
plt.show()
```

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html
from sklearn.cluster import KMeans
X = np.array([[1,2], [2,3], [4,2], [3,3], [-2,-2], [-1,-3], [-3,-4], [-2,-4]])
kmeans = KMeans(n_clusters=2).fit(X)

centers = kmeans.cluster_centers_
print(kmeans.predict(X))

plt.scatter(X[:,0], X[:,1])
plt.scatter(centers[:,0], centers[:,1])
plt.show()
```

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html
from sklearn.decomposition import PCA
X = np.array([[-1,-1], [-2,-1], [-3,-2], [1,1], [2,1], [3,2]])
pca = PCA(n_components=1).fit(X)

reduction = pca.transform(X)
reduction = (reduction-reduction.mean())/reduction.std()
print(reduction)
plt.scatter(X[:,0], X[:,1])
plt.scatter(reduction, [0]*len(X))
plt.show()
```

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.manifold.TSNE.html
from sklearn.manifold import TSNE
X = np.array([[-1, -1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]])
reduction = TSNE(n_components=1).fit_transform(X)
reduction = (reduction-reduction.mean())/reduction.std()

print(reduction)
plt.scatter(X[:,0], X[:,1])
plt.scatter(reduction, [0]*len(X))
plt.show()
```
