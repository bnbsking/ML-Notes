```python
import json, os, pickle

import cv2
import numpy as np
import matplotlib.pyplot as plt
import PIL
import yaml
```

### image

```python
# write
os.makedirs("_io", exist_ok=True)

img = np.array([[[255,0,0], [0,255,0]], [[0,0,255], [0,0,0]]], dtype=np.uint8)
# 1
cv2.imwrite("_io/img1.jpg", img)
# 2
plt.imshow(img)
plt.savefig("_io/img2.jpg")
plt.close()
# 3
PIL.Image.fromarray(img).save("_io/img3.jpg")
```

```python
# read
img1 = cv2.imread("_io/img1.jpg") # np.array
plt.imshow(img1)
plt.show()

img2 = plt.imread("_io/img2.jpg") # np.array
plt.imshow(img2)
plt.show()

img3 = np.array(PIL.Image.open("_io/img3.jpg")) # PIL -> np.array
plt.imshow(img3)
plt.show()
```

### config

```python
# save
D = {"a":1, "b":{"c":[2,3], "d":'4'}}

# 1
json.dump(D, open('./_io/data.json','w')) # cv2.imwrite place data at second, but others don't
# 2
yaml.dump(D, open('./_io/data.yaml','w'))
# 3
pickle.dump(D, open('./_io/data.pkl', 'wb'))
```

```python
# load
print( json.load(open('./_io/data.json', 'r')) )
print( yaml.safe_load(open('./_io/data.yaml', 'r')) )
print( pickle.load(open('./_io/data.pkl', 'rb')) )
```
