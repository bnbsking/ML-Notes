### Contour

```python
def f(x,y): # the height function
    return (1-x/2+x**5+y**3) * np.exp(-x**2-y**2)

x = np.linspace(-3,3,256)
y = np.linspace(-3,3,256)
X,Y = np.meshgrid(x, y)

plt.contourf(X, Y, f(X,Y), levels=8, alpha=0.75, cmap=plt.cm.hot) # face color # levels means number of colors
plt.colorbar()
contours = plt.contour(X, Y, f(X, Y), levels=8, colors="black")
plt.clabel(contours, fontsize=10)
plt.show()
```

### Confusion matrix

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

def confusion_matrix(M, class_list): # gt_pd same as sklearn
    n, M = len(M), np.array(M)
    R = np.array( [ [ ele/row.sum() if row.sum() else 0 for ele in row ] for row in M ]  ).round(3)
    P = np.array( [ [ ele/row.sum() if row.sum() else 0 for ele in row ] for row in M.T ] ).T.round(3)
    
    def f(matrix, title, subplot_idx, color_matrix=None):
        color_matrix = color_matrix if color_matrix is not None else matrix
        plt.subplot(*subplot_idx)
        plt.title(title, fontsize=14)
        plt.xlabel("PD")
        plt.ylabel("GT")
        plt.xticks(range(n), class_list, fontsize=14)
        plt.yticks(range(n), class_list, fontsize=14)
        plt.imshow(matrix, cmap=mpl.cm.Blues)
        plt.colorbar()
        for i in range(n):
            for j in range(n):
                plt.annotate(matrix[i][j], xy=(j,i), ha="center", va="center", \
                    color="black" if R[i][j]<0.5 else "white")    
    
    plt.figure(figsize=(18,4))
    f(M, "Number", (1,3,1))
    f(R, "Row-norm (Recall)", (1,3,2))
    f(P, "Col-norm (Precision)", (1,3,3))
    plt.show()

confusion_matrix([[2,0,1], [1,1,1], [0,1,1]], ["c0","c1","c2"])
```

### Statistical Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns

L = np.array([0,0,20,30,30,30])
kde = sns.kdeplot(L)
peak = kde.get_lines()[0].get_data()[1].max()
val, cnt = np.unique(L, return_counts=True)
plt.bar(val, cnt/cnt.max()*peak)
plt.show()
```
