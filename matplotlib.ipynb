```python
L1 = [1,3,2]
f = plt.figure(figsize=(8,4))
f.set_facecolor("#FF0000")

s1 = plt.subplot(1,2,1)
[x.set_linewidth(1.5) for x in s1.spines.values()] # border width
s1.plot(L1)
s1.set_facecolor("#FFFF00")

s2 = plt.subplot(1,2,2)
s1.plot(L1[::-1])
s2.plot(L1,color='green', marker='o', linestyle='dashed', linewidth=2, markersize=12)
for x,y in zip(range(3),L1):
    s2.annotate(y, xy=(x,y), fontsize=18)
    s2.text(x, y, y, ha="center", va="center", color="white", fontsize=12)
plt.xticks(np.linspace(0,2,3), ["A","B","C"])
    
plt.show()
```

### Pie

```python
L = [0.3, 0.2, 0.5]
plt.pie(L, labels=["blue", "orange", "green"], autopct="%.1f%%")
plt.show()
```

### bar

```python
students = ['Jack', 'Mary', 'Mike', 'David']
math = [78, 67, 90, 81]
history = [94, 71, 65, 88]
R, width = np.arange(0,len(students),1), 0.3 # width means bar
plt.bar(R+width/2, math, width, color='green', alpha=0.7) # centerx, y, barwidth
plt.bar(R-width/2, history, width, color='blue', alpha=0.7)
plt.xticks(range(len(students)), students)
plt.show()
```

```python
R, width = np.arange(0,4,1), 1 # width means bar
plt.bar(R, [0,1,2,3], width, color='red', alpha=0.5) # centerx, y, barwidth
plt.bar(R, [1,2,3,0], width, color='blue', alpha=0.5)
plt.show()
```

### Contour

```python
def f(x,y): # the height function
    return (1-x/2+x**5+y**3) * np.exp(-x**2-y**2)

x = np.linspace(-3,3,256)
y = np.linspace(-3,3,256)
X,Y = np.meshgrid(x, y)

plt.contourf(X, Y, f(X,Y), levels=8, alpha=0.75, cmap=plt.cm.hot) # face color # levels means number of colors
plt.colorbar()
contours = plt.contour(X, Y, f(X, Y), levels=8, colors="black")
plt.clabel(contours, fontsize=10)
plt.show()
```
